# **Matplotlib**

## **Introduction to Matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-darkgrid')
fig = plt.figure()
ax = plt.axes()
```

## **Multiple Plots**

```python
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

## **Colors and Styles**

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x-0), color="blue", linestyle="solid")
ax.plot(x, np.sin(x-1), color="g", linestyle="dashed")
ax.plot(x, np.sin(x-2), color=(0.3,0.6,0.7), linestyle="dotted")

# Shortcuts
ax.plot(x, np.sin(x-3),"-b")
ax.plot(x, np.sin(x-4), ":g")
ax.plot(x, np.sin(x-5), "-.k", linewidth=2)
ax.plot(x, np.sin(x-6), "--r", linewidth=3)
```
