## **Multiple Plots**

```python
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

## **Colors and Styles**

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x-0), color="blue", linestyle="solid")
ax.plot(x, np.sin(x-1), color="g", linestyle="dashed")
ax.plot(x, np.sin(x-2), color=(0.3,0.6,0.7), linestyle="dotted")

# Shortcuts
ax.plot(x, np.sin(x-3),"-b")
ax.plot(x, np.sin(x-4), ":g")
ax.plot(x, np.sin(x-5), "-.k", linewidth=2)
ax.plot(x, np.sin(x-6), "--r", linewidth=3)
```

## **Axis limits**

```python
plt.plot(x, np.arcsin(x), '-.r', linewidth=0.8)
plt.plot(10*x, np.cos(x), ':g', linewidth=2)
plt.xlim(0, 1)
plt.ylim(0, 1.4)
```

Limits shortcut

```python
plt.plot(x, np.arcsin(x), '-.b', linewidth=0.8)
plt.plot(2*x, np.cos(x), ':r', linewidth=2)
plt.axis([0,1,0,1.4])
```

Tight and equal options

- Tight will fit the graph in the corners
- Equal will not

```python
plt.plot(x, np.sin(x), '-.b', linewidth=0.8)
plt.plot(2*x, np.cos(x), ':r', linewidth=2)
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x), '-.b', linewidth=0.8)
plt.plot(2*x, np.cos(x), ':r', linewidth=2)
plt.axis('equal')
```

## **Legends label**

```python
plt.plot(x, np.sin(x), '-g', label="sin(x)")
plt.plot(x, np.cos(x), ':r', label="cos(x)")
plt.axis('equal')
plt.legend(loc="upper left") # It's necessary to call plt.legend()
plt.title("Sin-Cos-Curve")
plt.xlabel("X Values")
plt.ylabel("Y Values")
```

## **Set Function**

Equivalences between plt and ax
- plt.xlabel -> ax.set_xlabel
- plt.title -> ax.set_title
- plt.xlim -> ax.set_xlim

...

```python
x = np.linspace(0,100,10000)
ax = plt.axes()
ax.plot(x, np.sin(x), '-g', label="sin(x)")
ax.plot(x, np.cos(x), ':r', label="cos(x)")

ax.set(xlim=(1, 12), ylim=(-4, 5), xlabel="X Values", ylabel="Y Values", title="Sin-Cos-Curve" )

plt.legend()
```

## **Contour Plot**

We use plt.contourf() if we want a filled graphic or plot.contour if we don't.

- We need to pass as arguments our cartesian products of x and y, as well as a function to relate these two.

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)
X,Y = np.meshgrid(x, y)
Z = 2 * (X-5) * np.sin(Y)+2

plt.contourf(X, Y, Z, 50, cmap="coolwarm") # 50 is the interval
plt.colorbar()
```

np.meshgrid creates a cartesian product with the given arrays

## **Histograms**

```python
d1 = np.random.normal(0,0.7, 1000)
d2 = np.random.normal(-5, 2, 1000)
d3 = np.random.normal(5, 3, 1000)

plt.hist(d1, alpha=0.5, bins=50, histtype="stepfilled")
plt.hist(d2, alpha=0.5, bins=50, histtype="stepfilled")
plt.hist(d3, alpha=0.5, bins=50, histtype="stepfilled")
```

```python
params = dict(histtype='stepfilled', alpha=0.5, bins=30)

plt.hist(iris.sepal_length, **params)
plt.hist(iris.sepal_width, **params)
plt.hist(iris.petal_length, **params)
plt.hist(iris.petal_width, **params)
```

## **3D Plots**

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection='3d')
theta = np.linspace(-4*np.pi, 4*np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2+1
x = r*np.sin(theta) 
y = r*np.cos(theta)
ax.plot(x,y,z, label="3D Curve")
ax.legend()
```
