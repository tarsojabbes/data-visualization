# **Matplotlib**

## **Introduction to Matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-darkgrid')
fig = plt.figure()
ax = plt.axes()
```

## **Multiple Plots**

```python
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

## **Colors and Styles**

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x-0), color="blue", linestyle="solid")
ax.plot(x, np.sin(x-1), color="g", linestyle="dashed")
ax.plot(x, np.sin(x-2), color=(0.3,0.6,0.7), linestyle="dotted")

# Shortcuts
ax.plot(x, np.sin(x-3),"-b")
ax.plot(x, np.sin(x-4), ":g")
ax.plot(x, np.sin(x-5), "-.k", linewidth=2)
ax.plot(x, np.sin(x-6), "--r", linewidth=3)
```

## **Axis limits**

```python
plt.plot(x, np.arcsin(x), '-.r', linewidth=0.8)
plt.plot(10*x, np.cos(x), ':g', linewidth=2)
plt.xlim(0, 1)
plt.ylim(0, 1.4)
```

Limits shortcut

```python
plt.plot(x, np.arcsin(x), '-.b', linewidth=0.8)
plt.plot(2*x, np.cos(x), ':r', linewidth=2)
plt.axis([0,1,0,1.4])
```

Tight and equal options

- Tight will fit the graph in the corners
- Equal will not

```python
plt.plot(x, np.sin(x), '-.b', linewidth=0.8)
plt.plot(2*x, np.cos(x), ':r', linewidth=2)
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x), '-.b', linewidth=0.8)
plt.plot(2*x, np.cos(x), ':r', linewidth=2)
plt.axis('equal')
```

## **Legends label**

```python
plt.plot(x, np.sin(x), '-g', label="sin(x)")
plt.plot(x, np.cos(x), ':r', label="cos(x)")
plt.axis('equal')
plt.legend(loc="upper left") # It's necessary to call plt.legend()
plt.title("Sin-Cos-Curve")
plt.xlabel("X Values")
plt.ylabel("Y Values")
```

## **Set Function**

Equivalences between plt and ax
- plt.xlabel -> ax.set_xlabel
- plt.title -> ax.set_title
- plt.xlim -> ax.set_xlim

...

```python
x = np.linspace(0,100,10000)
ax = plt.axes()
ax.plot(x, np.sin(x), '-g', label="sin(x)")
ax.plot(x, np.cos(x), ':r', label="cos(x)")

ax.set(xlim=(1, 12), ylim=(-4, 5), xlabel="X Values", ylabel="Y Values", title="Sin-Cos-Curve" )

plt.legend()
```
