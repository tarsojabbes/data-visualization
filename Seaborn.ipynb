# **Seaborn**

## **Introduction to Seaborn**

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'x': np.random.normal(0,2,1000), 
                   'y': np.random.normal(3,5,1000),
                   'z': np.random.normal(-4,10,1000)})
```

```python
sns.set() # Sets the default style for Seaborn
for col in df.columns:
  plt.hist(df[col], density=True, alpha=0.5, bins=30)
  sns.kdeplot(df[col], shade=True)
```

```python
for col in df.columns:
  sns.distplot(df[col])
```

```python
with sns.axes_style('white'):
  sns.jointplot('x','y', df, kind="kde")
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.pairplot(iris, hue="species")
```

## **Relplot**

Scatter plots are called Relation plots, and we can set the hue to differenciate by species

```python
sns.relplot(x='sepal_length', y='sepal_width', data=iris, hue="species")
```

### **Kind Line**

```python
sns.relplot(x='sepal_length', y="sepal_width", 
            kind='line', data=iris, ci=None, 
            hue="species", style="species")
```

Hue set to species with set different colors to each specie, and style set to species will set different markers to each species (line, dashed, points etc)

ci can be None, 'sd' (standard deviation) and other values for our shadow 

### **Facets**

```python
tips = sns.load_dataset('tips')
```

```python
tips.time.unique()
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            col="time", data=tips)
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.subject.unique()
```

```python
sns.relplot(x="timepoint", y="signal", hue="event",
            style="event", col="subject", col_wrap=5,
            kind="line", data=fmri.query("region=='frontal'"))
```

## **Catplot**

```python
sns.set(style="ticks", color_codes=True)
```

With jitter set to False, our plots will look way more linear

```python
sns.catplot(x='day', y='total_bill', data=tips, jitter=False, hue="sex")
```
